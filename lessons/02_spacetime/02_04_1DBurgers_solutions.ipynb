```python
import numpy
import sympy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
from sympy import init_printing
init_printing()
```

```python
x = sympy.symbols('x')
```

```python
5*x
```

```python
x, nu, t = sympy.symbols('x nu t')
phi = sympy.exp(-(x-4*t)**2/(4*nu*(t+1)))+ sympy.exp(-(x-4*t-2*numpy.pi)**2/(4*nu*(t+1)))
phi
```

```python
phiprime = phi.diff(x)
phiprime
```

```python
from sympy.utilities.lambdify import lambdify

u = -2*nu*(phiprime/phi)+4
u
```

```python
u_lamb = lambdify((t,x,nu), u)
print("u at t=1, x=4, nu=3 is {}.".format(u_lamb(1,4,3)))
```

```python
# variables
nx = 101
nt = 100
dx = 2*numpy.pi/(nx-1)
nu = .07
sigma = .1
dt = sigma*dx**2/nu

x = numpy.linspace(0,2*numpy.pi, nx)
un = numpy.empty(nx)
t = 0
```

```python
u = numpy.asarray([u_lamb(t, x0, nu) for x0 in x])
u
```

```python
pyplot.figure(figsize=(8,5), dpi = 100)
pyplot.plot(x,u, ls='--', lw=3)
pyplot.xlim([0, 2*numpy.pi])
pyplot.ylim([0, 10]);
```

```python
for n in range(nt):
    un = u.copy()
    u[1:-1] = un[1:-1] - un[1:-1]*dt/dx*(un[1:-1] - un[:-2]) + nu*dt/dx**2*(un[2:]-2*un[1:-1]+un[:-2])
    u[0] = un[0]-un[0]*dt/dx*(un[0]-un[-1])+nu*dt/dx**2*(un[1] - 2*un[0]+un[-1])
    u[-1] = un[-1]-un[-1]*dt/dx*(un[-1] - un[-2]) + nu*dt/dx**2*(un[0]- 2*un[-1] + un[-2])
             
    u_analytical = numpy.asarray([u_lamb(nt*dt, xi, nu) for xi in x])         
```

```python
pyplot.figure(figsize=(8,5), dpi =100)
pyplot.plot(x, u, ls='--', lw=3, label='Computational')
pyplot.plot(x, u_analytical, label='Analytical')
pyplot.xlim([0, 2*numpy.pi])
pyplot.ylim([0,10])
pyplot.legend();
```

```python
fig = pyplot.figure(figsize=(8,6))
ax = pyplot.axes(xlim=(0,2*numpy.pi), ylim=(0,10))
line = ax.plot([],[], ls='--', lw=3)[0]
line2 = ax.plot([],[], 'k-', lw=2)[0]
ax.legend(['Computed', 'Analytical'])

def burgers(n):
    un = u.copy()
    
    u[1:-1] = un[1:-1] - un[1:-1] * dt/dx * (un[1:-1] - un[:-2]) + nu*dt/dx**2*\
                    (un[2:] - 2*un[1:-1] + un[:-2])

    u[0] = un[0] - un[0] * dt/dx * (un[0] - un[-1]) + nu*dt/dx**2*\
                (un[1] - 2*un[0] + un[-1])
    u[-1] = un[-1] - un[-1] * dt/dx * (un[-1] - un[-2]) + nu*dt/dx**2*\
                (un[0]- 2*un[-1] + un[-2])
        
    u_analytical = numpy.asarray([u_lamb(n*dt, xi, nu) for xi in x])
    line.set_data(x,u)
    line2.set_data(x, u_analytical)
    

animation.FuncAnimation(fig, burgers,
                        frames=nt, interval=100)
```
